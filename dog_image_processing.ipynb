How to resize a image using pillow library


```python
from PIL import Image   # importing pillow library
```

```python
img = Image.open('/content/dog1.jpg')
```

```python
img_resized = img.resize((200,200))
```

```python
img_resized.save('/content/dog_resize.jpg')
```

```python
# displaying the image from numpy array
img_re = mpimg.imread('/content/dog_resize.jpg')
img_plot = plt.imshow(img_re)
plt.show()
```

```python
img_re.shape
```

Converting RGB image to grayscale using openCV

```python
import cv2 # importing openCV library
```

```python
img=cv2.imread('/content/dog_resize.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscaleimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscaleimg)
```

```python
grayscaleimg.shape #RGB IMAGE IS COVERTED TO GRAYSCALE
```
